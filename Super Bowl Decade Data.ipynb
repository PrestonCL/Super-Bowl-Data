# Super Bowl Data by the Decade

## How many Super Bowls have there been?

```


df_2 = _deepnote_execute_sql("""SELECT COUNT(*) as Super_Bowl_Count
FROM 'super-bowl-ratings.csv'
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_2
```

## Networks and Viewership

What networks have hosted the Super Bowl? How often?

```


df_4 = _deepnote_execute_sql("""SELECT network, COUNT(*) as count
FROM 'super-bowl-ratings.csv'
GROUP BY network 
ORDER BY count 
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_4
```

What is the average viewership per network?

```


df_4 = _deepnote_execute_sql("""SELECT network, ROUND(avg(average_viewers), 0) as avg_viewers
FROM 'super-bowl-ratings.csv'
GROUP BY network
ORDER BY avg_viewers 
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_4
```

What is the average viewership per decade?

```


df_5 = _deepnote_execute_sql("""SELECT 
AVG(CASE 
    WHEN date BETWEEN '1960-01-01' AND '1969-12-31' THEN average_viewers  
    END) AS sixties,
AVG(CASE 
    WHEN date BETWEEN '1970-01-01' AND '1979-12-31' THEN average_viewers  
    END) AS seventies,
AVG(CASE 
    WHEN date BETWEEN '1980-01-01' AND '1989-12-31' THEN average_viewers  
    END) AS eighties,
AVG(CASE 
    WHEN date BETWEEN '1990-01-01' AND '1999-12-31' THEN average_viewers  
    END) AS ninties,
AVG(CASE 
    WHEN date BETWEEN '2000-01-01' AND '2009-12-31' THEN average_viewers  
    END) AS two_thousands,
AVG(CASE 
    WHEN date BETWEEN '2010-01-01' AND '2019-12-31' THEN average_viewers  
    END) AS twenty_tens,
AVG(CASE 
    WHEN date BETWEEN '2020-01-01' AND '2029-12-31' THEN average_viewers  
    END) AS twenty_twenties
FROM 'super-bowl-ratings.csv'
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_5
```

The top 5 watched Super Bowls

```


df_6 = _deepnote_execute_sql("""SELECT super_bowl_number, date, average_viewers
FROM 'super-bowl-ratings.csv'
ORDER BY average_viewers DESC
LIMIT 5

""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_6
```

The bottom 5 watched Super Bowls

```


df_7 = _deepnote_execute_sql("""SELECT super_bowl_number, date, average_viewers
FROM 'super-bowl-ratings.csv'
ORDER BY average_viewers ASC
LIMIT 5
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_7
```

## Super Bowl Ad Data

What is the average cost of a 30-second ad per decade?

```


df_9 = _deepnote_execute_sql("""SELECT 
AVG(CASE 
    WHEN date BETWEEN '1960-01-01' AND '1969-12-31' THEN cost_of_30_second_ad_usd
    END) AS sixties, 
AVG(CASE 
    WHEN date BETWEEN '1970-01-01' AND '1979-12-31' THEN cost_of_30_second_ad_usd  
    END) AS seventies,
AVG(CASE 
    WHEN date BETWEEN '1980-01-01' AND '1989-12-31' THEN cost_of_30_second_ad_usd 
    END) AS eighties,
AVG(CASE 
    WHEN date BETWEEN '1990-01-01' AND '1999-12-31' THEN cost_of_30_second_ad_usd 
    END) AS ninties,
AVG(CASE 
    WHEN date BETWEEN '2000-01-01' AND '2009-12-31' THEN cost_of_30_second_ad_usd
    END) AS two_thousands,
AVG(CASE 
    WHEN date BETWEEN '2010-01-01' AND '2019-12-31' THEN cost_of_30_second_ad_usd
    END) AS twenty_tens,
AVG(CASE 
    WHEN date BETWEEN '2020-01-01' AND '2029-12-31' THEN cost_of_30_second_ad_usd
    END) AS twenty_twenties
FROM 'super-bowl-ratings.csv'
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_9
```

How many viewers per dollar?

```


df_10 = _deepnote_execute_sql("""SELECT 
ROUND(AVG(CASE 
    WHEN date BETWEEN '1960-01-01' AND '1969-12-31' THEN average_viewers/CAST(cost_of_30_second_ad_usd AS DECIMAL)
    END), 0) AS sixties,
ROUND(AVG(CASE 
    WHEN date BETWEEN '1970-01-01' AND '1979-12-31' THEN average_viewers/CAST(cost_of_30_second_ad_usd AS DECIMAL)
    END), 0) AS seventies,
ROUND(AVG(CASE 
    WHEN date BETWEEN '1980-01-01' AND '1989-12-31' THEN average_viewers/CAST(cost_of_30_second_ad_usd AS DECIMAL)
    END), 0) AS eighties,
ROUND(AVG(CASE 
    WHEN date BETWEEN '1990-01-01' AND '1999-12-31' THEN average_viewers/CAST(cost_of_30_second_ad_usd AS DECIMAL)
    END), 0) AS ninties,
ROUND(AVG(CASE 
    WHEN date BETWEEN '2000-01-01' AND '2009-12-31' THEN average_viewers/CAST(cost_of_30_second_ad_usd AS DECIMAL)
    END), 0) AS two_thousands,
ROUND(AVG(CASE 
    WHEN date BETWEEN '2010-01-01' AND '2019-12-31' THEN average_viewers/CAST(cost_of_30_second_ad_usd AS DECIMAL)
    END), 0) AS twenty_tens,
ROUND(AVG(CASE 
    WHEN date BETWEEN '2020-01-01' AND '2029-12-31' THEN average_viewers/CAST(cost_of_30_second_ad_usd AS DECIMAL)
    END), 0) AS twenty_twenties
FROM 'super-bowl-ratings.csv'
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_10
```
